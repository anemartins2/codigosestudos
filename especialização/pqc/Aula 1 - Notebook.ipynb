## Síntese conceitual da aula

### A criptografia como infraestrutura invisível

A segurança da Internet moderna depende profundamente da criptografia de chave pública. Mecanismos como TLS, VPNs, autenticação de servidores, assinaturas digitais, certificados X.509 e cadeias de confiança baseiam-se em poucos problemas matemáticos bem estabelecidos: fatoração de inteiros e logaritmo discreto.

Esses mecanismos não são acessórios. Eles constituem a infraestrutura invisível da confiança digital, viabilizando comunicações seguras, comércio eletrônico, sistemas governamentais e serviços críticos.

### Computação quântica e a ruptura do paradigma clássico

A computação quântica altera radicalmente esse cenário ao introduzir algoritmos que exploram fenômenos físicos inexistentes no modelo clássico. 

O algoritmo de Shor resolve, em tempo polinomial, os problemas de fatoração e logaritmo discreto, quebrando RSA, Diffie–Hellman e ECC. Isso implica a quebra direta de:
- troca de chaves;
- assinaturas digitais;
- autenticação e não repúdio.

Por outro lado, o algoritmo de Grover afeta criptografia simétrica e funções hash apenas por um ganho quadrático, permitindo mitigação prática com chaves e digests maiores.

Essa assimetria estabelece um ponto central da aula: a ameaça quântica não afeta todos os mecanismos igualmente.

### Ameaça desbalanceada e impacto sistêmico

A consequência mais grave da computação quântica não é apenas a perda de confidencialidade, mas o colapso estrutural da confiança digital.

Quando algoritmos de chave pública se tornam quebráveis:
- certificados deixam de ser confiáveis;
- assinaturas podem ser forjadas;
- cadeias de confiança se rompem.

Diferentemente da criptografia simétrica, não existe ajuste incremental (como dobrar o tamanho da chave) que resolva o problema. A única alternativa é substituir os algoritmos.

### PQC × tecnologias quânticas

Diante dessa ameaça, surgem duas abordagens distintas:
- Criptografia Pós-Quântica (PQC):
    - algoritmos clássicos;
    - baseados em novos problemas matemáticos;
    - resistentes a adversários quânticos;
    - implantáveis sobre infraestrutura existente.
- Criptografia/comunicação quântica (QKD):
    - baseadas em fenômenos físicos quânticos;
    - exigem hardware especializado e canais dedicados;
    - aplicáveis a cenários restritos.

A PQC emerge como a estratégia principal de transição, enquanto tecnologias quânticas são complementares.

### *Store now, decrypt later*: um risco presente

A ameaça quântica não é apenas futura. Comunicações cifradas hoje podem ser capturadas, armazenadas e decifradas no futuro, quando computadores quânticos escaláveis estiverem disponíveis.

Esse modelo de ataque, conhecido como *store now, decrypt later*, torna vulneráveis informações com valor de longo prazo, como:
- segredos industriais;
- dados governamentais;
- informações médicas e biométricas;
- registros legais e financeiros.

A criptografia protege a longevidade da informação, não apenas o canal de comunicação atual.

### Impacto direto em TLS, PKI e assinaturas

A transição pós-quântica impacta diretamente:
- mecanismos de troca de chaves (TLS);
- certificados digitais e PKI;
- assinaturas de software e firmware;
- infraestruturas legais e institucionais.

Assinaturas digitais são particularmente críticas porque não admitem correção retroativa: uma assinatura forjada no futuro invalida documentos do passado. Isso torna a migração urgente e estratégica.

### Transição criptográfica como desafio de engenharia

Esta aula enfatiza que a transição pós-quântica não é apenas matemática, mas também:
- organizacional;
- econômica;
- regulatória;
- técnica.

A migração exige planejamento, testes, coexistência de algoritmos (modelos híbridos) e atualização de protocolos amplamente implantados.

### Mensagem central da aula

A computação quântica não ameaça apenas algoritmos isolados, mas a infraestrutura de confiança da Internet. A criptografia pós-quântica é necessária porque a segurança digital precisa sobreviver ao futuro.

## 1) Fatoração clássica de semiprimos

No RSA, um elemento central é o número
$$
n = p\cdot q,
$$
produto de dois primos grandes. A segurança do RSA clássico depende, em grande parte, da dificuldade de fatorar $n$ com recursos clássicos.

Aqui vamos fatorar semiprimos pequenos e observar como o custo cresce quando aumentamos o tamanho (em bits).

### O que observar

- Mesmo em tamanhos pequenos, o tempo tende a crescer rapidamente.
- A tendência é o que importa: em tamanhos reais, esse crescimento torna a fatoração impraticável.

Execute as células a seguir e observe a saída.

```python
def gen_semiprime(bits):
    """Gera n = p*q com p e q primos de aproximadamente bits/2."""
    half = bits // 2
    p = random_prime(2**half - 1, lbound=2^(half-1))
    q = random_prime(2**half - 1, lbound=2^(half-1))
    return p*q, p, q

def time_factor(n, repeats=1):
    """Mede o tempo para fatorar n usando factor(n)."""
    t_total = 0.0
    f = None
    for _ in range(repeats):
        t0 = time.time()
        f = factor(n)
        t_total += (time.time() - t0)
    return t_total / repeats, f

# Teste
n, p, q = gen_semiprime(32)
t, f = time_factor(n, repeats=1)
n, f, t
```

### 1.1 Tamanho (em bits) vs. tempo de fatoração

Execute a célula de código abaixo.

**Dicas para autoestudo:**
- Se seu computador for mais rápido/lento, os tempos absolutos variam.
- O ponto central é comparar **tendências** (como o tempo muda quando os bits aumentam).

Após executar, passe para a visualização simples (gráfico).


```python
bits_list = [24, 32, 40, 48, 56, 64]   # tamanhos pequenos para ilustração
trials = 5                                   # repetições para média (ajuste se desejar)

results = []  # (bits, tempo_medio_s)
for b in bits_list:
    times = []
    for _ in range(trials):
        N, p, q = gen_semiprime(b)
        t, _ = time_factor(N, repeats=1)
        times.append(t)
    results.append((b, float(np.mean(times))))

results
```

### 1.2 Visualização

O gráfico a seguir ajuda a enxergar a tendência do crescimento do tempo de fatoração.

Execute a célula e observe o formato da curva.

```python
import matplotlib.pyplot as plt

bits = [b for b, _ in results]
ts   = [t for _, t in results]

plt.figure()
plt.plot(bits, ts, marker='o')
plt.xlabel("Tamanho de N (bits)")
plt.ylabel("Tempo médio de fatoração (s)")
plt.title("Fatoração clássica: tempo vs. bits")
plt.grid(True)
plt.show()
```

## 2) Espaço de busca exponencial (força bruta) e chaves simétricas

Em criptografia simétrica (e em ataques contra certas construções de hash), uma ideia básica é a de **força bruta**.

Se uma chave tem $k$ bits, então existem $2^k$ chaves possíveis. Um atacante que tenta, “na tentativa e erro”, enfrenta um crescimento exponencial da dificuldade.

### O que observar
- Cada bit adicional dobra o espaço de busca.
- Isso motiva tamanhos mínimos de chave (por exemplo, 128 bits ou mais).

Execute a célula de código e compare os tempos estimados para diferentes tamanhos.

```python
def brute_force_time_seconds(k, rate_per_sec):
    """Tempo esperado ~ (2^(k-1)) / rate (metade do espaço em média)."""
    return (2**(k-1))/rate_per_sec

def human_time(seconds):
    """Formata segundos em unidades humanas."""    
    picosec  = 1e-12
    nanosec   = 1e-9
    microsec  = 1e-6
    milisec   = 1e-3
    sec = float(seconds)
    minute = 60.0
    hour   = 3600.0
    day    = 86400.0
    year   = 365.25 * day
    if sec < nanosec: return f"{sec/picosec:.2f} ps"
    if sec < microsec: return f"{sec/nanosec:.2f} ns"
    if sec < milisec: return f"{sec/microsec:.2f} μs"
    if sec < 1: return f"{sec/milisec:.2f} ms"
    if sec < minute: return f"{sec:.2f} s"
    if sec < hour:   return f"{sec/minute:.2f} min"
    if sec < day:    return f"{sec/hour:.2f} h"
    if sec < year:   return f"{sec/day:.2f} dias"
    return f"{sec/year:.2e} anos"

rates = {
    "1e6 tent/s": 1e6,
    "1e9 tent/s": 1e9,
    "1e12 tent/s": 1e12
}

k_list = [40, 56, 64, 80, 96, 128, 192, 256]

header = ["k (bits)"] + list(rates.keys())
table = []
for k in k_list:
    row = [k]
    for _, r in rates.items():
        row.append(human_time(brute_force_time_seconds(k, r)))
    table.append(row)

print("{:<10} {:>14} {:>14} {:>14}".format(*header))
for row in table:
    print("{:<10} {:>14} {:>14} {:>14}".format(*row))
```

## 3) Intuição para Grover

Grover é um algoritmo quântico para busca não estruturada que oferece aceleração quadrática.

De forma simplificada:
- Se um ataque clássico custa $2^k$,
- um ataque quântico com Grover custa aproximadamente $2^{k/2}$.

Isso não “quebra” algoritmos simétricos como Shor quebra RSA/ECC, mas reduz a margem de segurança.

Execute a célula e observe a regra prática de equivalência $k \rightarrow k/2$).

```python
def grover_effective_security_bits(k):
    """Aproximação padrão: Grover reduz segurança efetiva para k/2 bits."""
    return k/2

for k in [40, 56, 64, 80, 96, 128, 192, 256]:
    print(f"AES-{k}: segurança clássica ~ {k} bits; com Grover ~ {grover_effective_security_bits(k):.0f} bits")
```

## 4) Comparação conceitual de crescimento

Nesta seção, o objetivo não é obter “custos exatos”, mas comparar formas de crescimento:

- **Busca exaustiva:** exponencial $2^k$ (e com Grover $\sim 2^{k/2}$).
- **Fatoração (clássica):** subexponencial (por exemplo, GNFS).
- **Impacto de Shor:** torna fatoração/log discreto eficientes em um modelo quântico ideal.

> Obs.: GNFS é a sigla de *General Number Field Sieve*, o algoritmo clássico mais eficiente conhecido para fatoração de inteiros grandes. Ele é o principal responsável pelo fato da fatoração não ter crescimento exponencial clássico, mas sim subexponencial.

Execute a célula para ver uma comparação de tendências e use-a para consolidar a mensagem da aula.

```python
import math

def gnfs_rough_cost(n_bits):
    # Exp( c*(ln N)^(1/3)*(ln ln N)^(2/3) ), com c = (64/9)^(1/3). Escala ilustrativa.
    lnN = n_bits * math.log(2.0)
    lnlnN = math.log(lnN)
    c = (64.0/9.0)**(1.0/3.0)
    return math.exp(c * (lnN**(1.0/3.0)) * (lnlnN**(2.0/3.0)))

def shor_poly_cost(n_bits):
    return n_bits**3  # escala ilustrativa

for n in [256, 512, 1024, 2048, 3072]:
    print(f"n={n:4d} bits | GNFS~ {gnfs_rough_cost(n):.3e} (escala) | Shor~ {shor_poly_cost(n):.3e} (escala)")
```
