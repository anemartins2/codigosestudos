```python
def brute_force_time_seconds(k, rate_per_sec):
    """Tempo esperado ~ (2^(k-1)) / rate (metade do espaço em média)."""
    return (2^(k-1)) / rate_per_sec

def human_time(seconds):
    """Formata segundos em unidades humanas."""    
    picosec  = 1e-12
    nanosec   = 1e-9
    microsec  = 1e-6
    milisec   = 1e-3
    sec = float(seconds)
    minute = 60.0
    hour   = 3600.0
    day    = 86400.0
    year   = 365.25 * day
    if sec < nanosec: return f"{sec/picosec:.2f} ps"
    if sec < microsec: return f"{sec/nanosec:.2f} ns"
    if sec < milisec: return f"{sec/microsec:.2f} μs"
    if sec < 1: return f"{sec/milisec:.2f} ms"
    if sec < minute: return f"{sec:.2f} s"
    if sec < hour:   return f"{sec/minute:.2f} min"
    if sec < day:    return f"{sec/hour:.2f} h"
    if sec < year:   return f"{sec/day:.2f} dias"
    return f"{sec/year:.2e} anos"

rates = {
    "1e6 tent/s": 1e6,
    "1e9 tent/s": 1e9,
    "1e12 tent/s": 1e12
}

k_list = [40, 56, 64, 80, 96, 128, 192, 256]

header = ["k (bits)"] + list(rates.keys())
table = []
for k in k_list:
    row = [k]
    for _, r in rates.items():
        row.append(human_time(brute_force_time_seconds(k, r)))
    table.append(row)

print("{:<10} {:>14} {:>14} {:>14}".format(*header))
for row in table:
    print("{:<10} {:>14} {:>14} {:>14}".format(*row))
```
