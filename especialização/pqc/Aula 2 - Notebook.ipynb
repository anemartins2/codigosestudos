## Síntese conceitual da aula

### Reticulados como base estrutural da PQC

A criptografia pós-quântica moderna fundamenta-se majoritariamente em reticulados euclidianos de alta dimensão, estruturas discretas e regulares que emergem de combinações inteiras de vetores base. A segurança decorre não de uma operação algébrica específica, mas da dificuldade geométrica intrínseca associada a esses objetos, especialmente quando descritos por bases “ruins”, com vetores longos e altamente correlacionados.

A mesma estrutura geométrica pode admitir múltiplas bases: algumas tornam evidentes vetores curtos (bases “boas”), enquanto outras ocultam completamente essas direções privilegiadas (bases “ruins”). Ataques criptográficos, nesse contexto, equivalem à busca por uma base boa, um problema computacionalmente intratável em alta dimensão.

### Problemas geométricos centrais: SVP e CVP

Dois problemas fundamentais governam a complexidade dos reticulados:
- SVP (Shortest Vector Problem): encontrar o menor vetor não nulo de um reticulado;
- CVP (Closest Vector Problem): dado um ponto arbitrário, encontrar o vetor do reticulado mais próximo.

Ambos são NP-difíceis e permanecem difíceis mesmo quando se aceita apenas uma aproximação da solução, sobretudo em espaços de centenas ou milhares de dimensões. Esses problemas modelam, de forma abstrata, tarefas de decodificação sob ruído, que é o cenário explorado pela criptografia baseada em LWE.

### Alta dimensão e “maldição geométrica”

Os reticulados utilizados em PQC não vivem em 2D ou 3D, mas em espaços de dimensão muito elevada. Nesse regime:
- vetores curtos tornam-se raros;
- ângulos tendem à ortogonalidade;
- aproximações locais deixam de fornecer informação global.

Essa “maldição da dimensionalidade” explica por que intuições geométricas simples continuam válidas, mas algoritmos geométricos eficientes deixam de existir. É justamente esse descompasso que sustenta a segurança prática da PQC.

### LWE: da álgebra linear ao problema criptográfico

O problema *Learning With Errors* (LWE) surge quando um sistema linear modular simples
$$
A x = b \pmod q
$$
é perturbado por um ruído pequeno e controlado, transformando-se em
$$
A x \approx b \pmod q.
$$

Sem ruído, o sistema é facilmente resolvido. Com ruído, a linearidade exata é quebrada, e o problema passa a ser geometricamente equivalente a um CVP ruidoso em alta dimensão. O segredo deixa de ser recuperável por métodos algébricos clássicos, mesmo quando o ruído é pequeno.

Um aspecto crucial é que o LWE é difícil em instância média: não apenas casos artificiais ou patológicos são difíceis, mas instâncias aleatórias típicas, o que é essencial para segurança criptográfica.

### Ruído: erro aparente, elemento central de segurança

Na PQC baseada em LWE, o ruído não é um defeito, mas o mecanismo central de segurança:
- ruído inexistente → sistema linear resolúvel;
- ruído excessivo → falha de correção entre partes legítimas;
- ruído bem calibrado → simultaneamente correção e segurança.

Esse equilíbrio explica por que o controle estatístico do erro é um dos principais desafios de engenharia criptográfica em esquemas pós-quânticos.

### RLWE e MLWE: estruturando o LWE para uso prático

O LWE “puro” é conceitualmente robusto, mas ineficiente. Para viabilizar uso real, introduzem-se estruturas algébricas controladas:
- RLWE (Ring-LWE): impõe estrutura de anel, permitindo operações rápidas (NTT) e chaves menores, ao custo de maior cautela criptográfica;
- MLWE (Module-LWE): compromisso entre LWE e RLWE, oferecendo flexibilidade de parâmetros, eficiência prática e maior conservadorismo estrutural.

Os principais padrões do NIST (Kyber/ML-KEM e Dilithium/ML-DSA) baseiam-se em MLWE, justamente por equilibrar segurança teórica e desempenho para a Internet.

### Da matemática abstrata à criptografia utilizável

A aula conclui enfatizando que matemática difícil não basta. Para transformar LWE em criptografia real, são necessários:
- algoritmos completos (KeyGen, Encaps/Decaps, Sign/Verify);
- mecanismos auxiliares (amostragem, compressão, reconciliação);
- garantias formais de correção, segurança e eficiência.

### Mensagem central da aula

O ruído transforma um problema linear fácil em um problema computacionalmente difícil, ao mesmo tempo em que permite correção para as partes legítimas. Essa ambiguidade controlada é a base da segurança do LWE e do MLWE.

## Preparação

Na célula a seguir, você apenas carrega bibliotecas, definir parâmetros globais e funções:
- `small_noise_vec(n, bound)`: gera um vetor de ruído pequeno (por exemplo, em $\{-1,0,1\}^n$).
- `lift_centered(x, q)`: converte valores módulo $q$ para inteiros “centrados” no intervalo $( -q/2,\; q/2)$.

O objetivo desses funções é facilitar a interpretação dos resultados e evitar que valores como $q-1$ apareçam como “grandes”, quando na verdade representam $-1$ módulo $q$.

Essas funções não são o foco do notebook. Você pode tratá-las como “caixas-pretas” utilitárias.

```python
# Notebook para kernel SageMath.
from sage.all import *
import random

def small_noise_vec(n, bound=1):
    """Retorna vetor de ruído pequeno em ZZ: cada coordenada em [-bound, bound]."""
    return vector(ZZ, [ZZ.random_element(-bound, bound+1) for _ in range(n)])

def invertible_matrix_mod_q(n, q):
    """Gera matriz n×n invertível em GF(q)."""
    F = GF(q)
    while True:
        A = random_matrix(F, n, n)
        if A.is_invertible():
            return A

def lift_centered(v, q):
    """Levanta elementos mod q para inteiros centrados em (-q/2, q/2]."""
    out = []
    for x in v:
        xi = ZZ(int(x)) % q
        if xi > q//2:
            xi -= q
        out.append(xi)
    return vector(ZZ, out)
```

## 1) Sistema linear sem ruído

Aqui simulamos o cenário mais simples:

$$
b = A s \pmod{q},
$$

sem qualquer perturbação. Em seguida, tentamos recuperar $s$ resolvendo o sistema em $\mathbb{Z}_q$.

### O que observar

Se não há ruído, o problema se comporta como um sistema linear comum (apenas com aritmética módulo $q$), e a recuperação tende a funcionar.

```python
# Parâmetros didáticos
q = 101
n = 4
F = GF(q)

A = invertible_matrix_mod_q(n, q)
s = random_vector(F, n)          # segredo em GF(q)
b = A * s                        # sem ruído

s_rec = A.solve_right(b)

print("q =", q, "| n =", n)
print("Segredo s     =", s)
print("Recuperado s' =", s_rec)
print("Recuperou corretamente?", s_rec == s)
```

### Discussão

**Por que** um sistema linear exato (sem ruído), mesmo módulo $q$, tende a ser tratável em computadores clássicos?  

Ao executar o código, observe como a estrutura “exata” do sistema torna a recuperação possível.

## 2) Sistema linear com ruído

Agora introduzimos ruído pequeno:

$$
b = A s + e \pmod{q},
$$

onde $e$ é um vetor com valores pequenos.

### O que esperar

Mesmo ruído pequeno pode impedir a recuperação direta do segredo, pois o sistema deixa de ser “exato”.

Execute a célula e observe se o método de recuperação ainda funciona.

```python
eZZ = small_noise_vec(n, bound=1)  # ruído em {-1,0,1}^n
eF  = vector(F, [F(int(x)) for x in eZZ])

b_noisy = b + eF
s_rec_noisy = A.solve_right(b_noisy)

print("Ruído e (ZZ)         =", eZZ)
print("b (sem ruído)        =", b)
print("b_noisy (com ruído)  =", b_noisy)
print("Segredo s            =", s)
print("Recuperado com ruído =", s_rec_noisy)
print("Recuperou corretamente?", s_rec_noisy == s)

delta = s_rec_noisy - s
print("Delta = s' - s (mod q) =", delta)
```

## 3) Intuição para LWE com muitas amostras

Nesta parte simulamos o cenário típico do LWE:

- uma matriz $A$ aleatória,
- um segredo $s$,
- ruído pequeno,
- e muitas amostras (muitas equações).

A pergunta é natural: *“Se eu tiver amostras suficientes, consigo anular o efeito do ruído e recuperar o segredo?”*  

As próximas seções mostram por que essa intuição falha.

```python
# Parâmetros didáticos de LWE
q = 97
n = 4
m = 14
F = GF(q)

A = random_matrix(F, m, n)
s = random_vector(F, n)

eZZ = small_noise_vec(m, bound=1)
eF  = vector(F, [F(int(x)) for x in eZZ])

b = A*s + eF

print("q =", q, "| n =", n, "| m =", m)
print("Segredo s =", s)
print("Amostra de ruído eZZ[0:8] =", list(eZZ)[:8])
```

### 3.1 Tentativa de recuperação ingênua por ajuste linear

Nesta seção fazemos uma tentativa propositalmente ingênua: tratamos o problema como se fosse apenas um ajuste linear em números reais.

A ideia é:

1. Converter os dados (módulo $q$) para uma forma “centrada” em inteiros, para facilitar leitura.
2. Procurar um vetor $\hat{s}$ que produza valores o mais próximos possível de $b$ (no sentido usual de “melhor ajuste”).
3. Arredondar a solução para obter uma estimativa inteira e comparar com o segredo real.

O código consegue produzir uma estimativa $\hat{s}$, mas ela tipicamente não coincide com o segredo real. Isso não é um defeito do notebook; é um sinal de que LWE não cede a métodos lineares simples.

Para garantir que o cálculo sempre seja possível, o ajuste inclui um termo pequeno de estabilização numérica. Você não precisa dominar isso para entender o experimento.

```python
# 3.1 Recuperação ingênua (com regularização ridge)
#
# Em vez de usar (A^T A)^{-1} A^T b (que frequentemente falha por A^T A ser singular),
# usamos a versão regularizada:
#     x_hat = (A^T A + λ I)^{-1} A^T b
# Isso garante invertibilidade e permite calcular um s_hat para fins de intuição.

# Levanta A e b para ZZ centrado
A_ZZ = matrix(ZZ, [[lift_centered([A[i,j]], q)[0] for j in range(n)] for i in range(m)])
b_ZZ = lift_centered(b, q)

# Converte para RR (least squares é naturalmente um problema em números reais)
A_RR = matrix(RR, A_ZZ)
b_RR = vector(RR, list(b_ZZ))

ncols = A_RR.ncols()
lam = 1e-3   # ajuste pequeno; pode testar 1e-4 ou 1e-2

AtA = A_RR.transpose() * A_RR
Atb = A_RR.transpose() * b_RR

# Solução ridge (bem definida para λ>0)
x_hat = (AtA + lam*identity_matrix(RR, ncols)).solve_right(Atb)

# Arredonda para ZZ e reduz mod q para obter uma "estimativa" do segredo
x_round = vector(ZZ, [ZZ(round(rr)) for rr in x_hat])
s_hat = vector(F, [F(int(xx)) for xx in x_round])

print("λ (ridge)                =", lam)
print("x_hat (RR)               =", x_hat)
print("x_round (ZZ)             =", x_round)
print("Estimativa s_hat (mod q) =", s_hat)
print("Acertou?", s_hat == s)
```

### 3.2 Resíduo

Mesmo que $\hat{s}$ não seja o segredo, podemos medir quão ruim foi o ajuste.

Para isso, calculamos o resíduo (erro que sobra):

$$
r = A\hat{s} - b,
$$

avaliado em inteiros “centrados” (para interpretar corretamente a aritmética módulo $q$).

### O que observar

- Se o ajuste fosse bom, o resíduo seria pequeno.
- No LWE, o resíduo tende a permanecer grande: o ajuste linear não “explica” os dados.

```python
def mod_centered_matvec(A_F, s_F, q):
    v = A_F * s_F
    return lift_centered(v, q)

# Resíduo centrado r = A*s_hat - b (tudo em ZZ centrado)
pred = mod_centered_matvec(A, s_hat, q)
resid = pred - b_ZZ

l2 = sqrt(sum([rr^2 for rr in resid]))
linf = max([abs(rr) for rr in resid])

print("||resíduo||_2   =", l2)
print("||resíduo||_inf =", linf)

# Dica de interpretação:
# - Se o ruído é pequeno, o resíduo tende a refletir principalmente o erro e.
# - Mesmo assim, isso NÃO recupera o segredo s de forma confiável no cenário modular.
```
