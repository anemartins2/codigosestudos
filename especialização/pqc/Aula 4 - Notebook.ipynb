## Síntese conceitual da aula

### O papel crítico das assinaturas digitais

Assinaturas digitais são o alicerce da confiança digital. Elas garantem autenticidade, integridade e não repúdio, sustentando:
- TLS e autenticação de servidores;
- Infraestruturas de Chave Pública (PKI);
- Atualizações seguras de software e firmware;
- Cadeias de confiança institucionais e legais.

Diferentemente da confidencialidade, assinaturas exigem validade de longo prazo: uma assinatura forjada no futuro invalida documentos, certificados e registros criados no passado.

### A ameaça quântica às assinaturas clássicas

O algoritmo de Shor quebra diretamente RSA e ECDSA, permitindo:
- falsificação de assinaturas;
- colapso de cadeias de confiança;
- perda de validade de certificados e documentos históricos.

Não existe “correção retroativa” para assinaturas digitais. Uma vez que o esquema é quebrado, todas as assinaturas passadas tornam-se questionáveis. Por isso, assinaturas digitais são um dos vetores mais críticos da transição pós-quântica.

### Requisitos de assinaturas pós-quânticas

Assinaturas digitais pós-quânticas devem satisfazer simultaneamente:
- correção;
- verificabilidade pública;
- resistência à falsificação;
- segurança formal EUF-CMA (mesmo sob ataques adaptativos);
- robustez de implementação;
- validade criptográfica por décadas.

Esses requisitos impõem uma combinação cuidadosa de hipóteses matemáticas sólidas e engenharia criptográfica rigorosa.

### Assinaturas baseadas em reticulados: ML-DSA e FN-DSA

Assim como nos KEMs, reticulados fornecem problemas difíceis mesmo para adversários quânticos.

O ML-DSA (Dilithium) baseia-se em:
- MLWE, para proteger a chave secreta;
- MSIS, para impedir falsificação.

A assinatura é construída como uma relação algébrica ruidosa, envolvendo:
- hash da mensagem (desafio imprevisível);
- ruído controlado (proteção da chave);
- rejection sampling (evita vazamento estatístico).

O resultado é um esquema robusto, equilibrado e seguro por projeto, sendo o padrão preferencial do NIST.

já o FN-DSA (Falcon) utiliza reticulados NTRU e o algoritmo GPV, tratando assinatura como amostragem de uma pré-imagem curta de uma equação pública.

A complexidade do Falcon não está na verificação, mas na amostragem gaussiana extremamente precisa, que:
- produz assinaturas muito compactas;
- exige implementação numérica cuidadosa;
- torna o esquema sensível a erros de engenharia.

O Falcon oferece máxima eficiência, ao custo de maior complexidade de implementação.

### Assinaturas baseadas em hash: SLH-DSA (SPHINCS+)

O SPHINCS+ adota a abordagem oposta:
- nenhuma álgebra estruturada;
- segurança reduzida exclusivamente às propriedades das funções hash.

Ele organiza assinaturas em hiperárvores de Merkle, combinando:
- FORS (assinatura eficiente de mensagens);
- WOTS+ (assinatura de uso único baseada em hash).

O preço dessa cautela criptográfica é:
- assinaturas grandes;
- maior custo computacional.

O SPHINCS+ é usado quando se deseja hipóteses mínimas, mesmo sacrificando eficiência.

### Comparação e adoção prática

Os esquemas pós-quânticos de assinatura formam um espectro claro:
- ML-DSA (Dilithium) → equilíbrio entre segurança, desempenho e robustez;
- FN-DSA (Falcon) → eficiência extrema, maior complexidade;
- SLH-DSA (SPHINCS+) → segurança conservadora, custo elevado.

A adoção no mundo real envolve:
- migração gradual;
- esquemas híbridos;
- avaliação contínua de desempenho e maturidade;
- impacto institucional de longo prazo.

Decisões tomadas hoje afetam a validade criptográfica por décadas.

### Mensagem central da aula

Assinaturas digitais pós-quânticas não protegem apenas dados, mas a confiança, a validade jurídica e a memória digital de longo prazo. Em um mundo quântico, a escolha do esquema define o futuro da confiança digital.

## 1) Estrutura geral de assinatura digital

Uma assinatura digital oferece, em termos práticos:

- **Autenticidade**: foi quem assinou?
- **Integridade**: a mensagem foi alterada?
- **Não repúdio** (no sentido criptográfico): a assinatura vincula o signatário à mensagem, dadas as suposições do sistema.

Nesta seção, você verá o fluxo mínimo:

- `Sign(sk, msg) -> sig`
- `Verify(pk, msg, sig) -> True/False`

Execute as células a seguir e observe que:
- a verificação passa para a mensagem original;  
- a verificação falha se a mensagem for alterada.

```python
def H_bytes(data: bytes) -> bytes:
    return hashlib.sha256(data).digest()

def H_int_mod_q(msg: str, q: int) -> int:
    h = H_bytes(msg.encode('utf-8'))
    return int.from_bytes(h, 'big') % q

# Toy signature A (NÃO seguro):
# KeyGen: sk = x in Z_q^*, pk = x^{-1} mod q
# Sign:  σ = x * H(m) mod q
# Verify: pk * σ == H(m) mod q

def toy_sig_keygen(q=65537):
    while True:
        x = ZZ.random_element(2, q-1)
        if gcd(x, q) == 1:
            break
    xinv = inverse_mod(x, q)
    return xinv, x  # (pk, sk)

def toy_sig_sign(sk, msg, q=65537):
    h = H_int_mod_q(msg, q)
    return (sk * h) % q

def toy_sig_verify(pk, msg, sig, q=65537):
    h = H_int_mod_q(msg, q)
    return (pk * sig) % q == h % q

q = 65537
pk, sk = toy_sig_keygen(q=q)
m = "Mensagem de teste"
sig = toy_sig_sign(sk, m, q=q)
print("Verificação OK?", toy_sig_verify(pk, m, sig, q=q))
print("Verifica msg adulterada?", toy_sig_verify(pk, m+'X', sig, q=q))
```

## 2) Intuição para assinaturas baseadas em reticulados

Assinaturas baseadas em reticulados (como ML-DSA e FN-DSA) utilizam, em alto nível, estruturas algébricas eficientes e ruído controlado para garantir segurança.

O código desta seção não implementa ML-DSA e FN-DSA reais, mas ajuda a enxergar a ideia de que:
- existe um “limiar” (tolerância);
- pequenas variações (ruído) podem levar à aceitação ou rejeição.

Execute as células e observe como o parâmetro de ruído (ou tolerância) afeta o comportamento.

```python
q = 97
n = 4
F = GF(q)

def small_noise_vec_ZZ(n, bound=1):
    return vector(ZZ, [ZZ.random_element(-bound, bound+1) for _ in range(n)])

def H_challenge_bit(u_vec, msg, q):
    data = (str(list(u_vec)) + "|" + msg).encode('utf-8')
    h = hashlib.sha256(data).digest()
    return int.from_bytes(h, 'big') & 1

def lattice_toy_keygen(q=97, n=4, bound=1):
    F = GF(q)
    A = random_matrix(F, n, n)
    s = random_vector(F, n)
    e = vector(F, [F(int(x)) for x in small_noise_vec_ZZ(n, bound)])
    t = A*s + e
    return (A, t), s

def lattice_toy_sign(sk, pk, msg, y_bound=3):
    (A, t) = pk
    q = A.base_ring().order()
    F = A.base_ring()
    n = A.nrows()

    yZZ = small_noise_vec_ZZ(n, bound=y_bound)
    y = vector(F, [F(int(x)) for x in yZZ])

    u = A*y
    c = H_challenge_bit(u, msg, q)   # 0/1
    z = y + (c * sk)
    return (u, z, c), yZZ

def lattice_toy_verify(pk, msg, sig, tol=2):
    (A, t) = pk
    q = A.base_ring().order()
    (u, z, c) = sig

    lhs = A*z - u - (c * t)

    # levanta para inteiros centrados
    lhsZZ = []
    for x in lhs:
        xi = ZZ(int(x)) % q
        if xi > q//2: xi -= q
        lhsZZ.append(xi)
    lhsZZ = vector(ZZ, lhsZZ)

    ok = all(abs(int(v)) <= tol for v in lhsZZ)
    return ok, lhsZZ

pk, sk = lattice_toy_keygen(q=q, n=n, bound=1)
msg = "Assinar esta mensagem"
sig, _ = lattice_toy_sign(sk, pk, msg, y_bound=3)
ok, resid = lattice_toy_verify(pk, msg, sig, tol=2)
print("Verificação OK?", ok, "| resíduo:", resid)

ok2, resid2 = lattice_toy_verify(pk, msg+'X', sig, tol=2)
print("Verifica msg adulterada?", ok2, "| resíduo:", resid2)
```

### 2.1 Taxa de sucesso e efeito do ruído/limiar

Nesta etapa, você mede empiricamente a taxa de verificação bem-sucedida quando o ruído ou o limiar variam.

Em esquemas reais, a taxa de falha deve ser extremamente baixa. Parâmetros precisamser ajustados para equilibrar a correção (aceitar assinaturas válidas) e a segurança.

Execute a célula e registre a taxa de sucesso.

```python
def lattice_toy_success(trials=300, y_bound=3, tol=2, key_noise_bound=1):
    ok = 0
    for _ in range(trials):
        pk, sk = lattice_toy_keygen(q=q, n=n, bound=key_noise_bound)
        msg = "m" + str(ZZ.random_element(0, 10^9))
        sig, _ = lattice_toy_sign(sk, pk, msg, y_bound=y_bound)
        ok_i, _ = lattice_toy_verify(pk, msg, sig, tol=tol)
        ok += 1 if ok_i else 0
    return ok/trials

for yb in [2, 3, 4, 5]:
    print("y_bound=", yb, "-> sucesso=", lattice_toy_success(trials=300, y_bound=yb, tol=2, key_noise_bound=1))
```

## 3) Intuição para SLH-DSA (SPHINCS+)

Assinaturas hash-based, como SLH-DSA, são construídas a partir de:
- funções hash criptográficas,
- assinaturas de uso único (em nível conceitual),
- e uma árvore de Merkle para autenticar muitas chaves/folhas.

A intuição central é que, para verificar uma assinatura, você precisa de um caminho de autenticação (*auth path*), cujo tamanho cresce aproximadamente como $\log_2(N)$, em que $N$ é o número de folhas.

Execute as células e observe como o *auth path* é usado para reconstruir a raiz e validar a assinatura.

```python
def sha256(b: bytes) -> bytes:
    return hashlib.sha256(b).digest()

def merkle_parent(a: bytes, b: bytes) -> bytes:
    return sha256(a + b)

def build_merkle_root(leaves):
    level = leaves[:]
    while len(level) > 1:
        nxt = []
        for i in range(0, len(level), 2):
            nxt.append(merkle_parent(level[i], level[i+1]))
        level = nxt
    return level[0]

def merkle_auth_path(leaves, idx):
    path = []
    level = leaves[:]
    i = idx
    while len(level) > 1:
        sibling = i ^ 1
        path.append(level[sibling])
        nxt = []
        for j in range(0, len(level), 2):
            nxt.append(merkle_parent(level[j], level[j+1]))
        level = nxt
        i //= 2
    return path

def merkle_verify_leaf(leaf, idx, path, root):
    x = leaf
    i = idx
    for sib in path:
        if i % 2 == 0:
            x = merkle_parent(x, sib)
        else:
            x = merkle_parent(sib, x)
        i //= 2
    return x == root

# OTS toy: pk = H(sk), assinatura = sk (revelado). Verificação: H(sig)==pk
def ots_keygen():
    sk = os.urandom(32)
    pk = sha256(sk)
    return pk, sk

def ots_sign(sk, msg: str):
    return sk

def ots_verify(pk, msg: str, sig):
    return sha256(sig) == pk

# Monta uma Merkle com N folhas (potência de 2)
N = 8
ots_pks, ots_sks = [], []
for _ in range(N):
    pk_i, sk_i = ots_keygen()
    ots_pks.append(pk_i); ots_sks.append(sk_i)

leaves = [sha256(pk_i) for pk_i in ots_pks]
root = build_merkle_root(leaves)

print("Merkle root (pk) =", root.hex()[:32], "...")
```

### 3.1 Assinar e verificar

Aqui você executa um ciclo completo:
1. gerar folhas e a raiz da árvore;
2. assinar uma mensagem em uma folha;
3. verificar reconstruindo a raiz a partir do caminho de autenticação.

Execute a célula e confirme que:
- a assinatura verifica para a mensagem original;
- falha se a mensagem for modificada.

```python
def hashbased_toy_sign(msg: str, i: int, ots_sks, leaves):
    sig_ots = ots_sign(ots_sks[i], msg)
    auth = merkle_auth_path(leaves, i)
    return (i, sig_ots, auth)

def hashbased_toy_verify(msg: str, signature, root, ots_pks):
    i, sig_ots, auth = signature
    if not ots_verify(ots_pks[i], msg, sig_ots):
        return False
    leaf = sha256(ots_pks[i])
    return merkle_verify_leaf(leaf, i, auth, root)

msg = "Mensagem para assinar (hash-based toy)"
i = 3
sig = hashbased_toy_sign(msg, i, ots_sks, leaves)
ok = hashbased_toy_verify(msg, sig, root, ots_pks)

print("Verificação OK?", ok)
print("Tamanho sig_ots (bytes):", len(sig[1]))
print("auth_path:", len(sig[2]), "nós (cada um 32 bytes)")
print("Tamanho aprox assinatura (bytes):", len(sig[1]) + 32*len(sig[2]) + 4)
```

## 4) Comparação

Compare, em termos conceituais:
- **Lattice-based (ML-DSA / FN-DSA):**
  - muito eficientes,
  - dependem de hipóteses ligadas a reticulados e ruído,
  - implementação pode exigir cuidado (especialmente FN-DSA/FALCON em ambientes reais).

- **Hash-based (SLH-DSA):**
  - depende principalmente da segurança de funções hash,
  - tende a ter assinaturas maiores e custo de verificação/geração distinto,
  - é uma alternativa robusta quando se quer suposições mais simples.
