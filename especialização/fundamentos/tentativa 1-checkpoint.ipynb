## Algoritmo de grover sem ruido 


```python

from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, array_to_latex
from IPython.display import display, Markdown, Latex
import matplotlib.pyplot as plt
import numpy as np

```

```python
# =====================================================
# ALGORITMO DE GROVER - VERSÃO SIMPLES E DIDÁTICA
# =====================================================
# Objetivo: encontrar o estado |11⟩ em um banco de 4 itens
# Usa 2 qubits + 1 iteração (número ótimo para 4 itens)

from qiskit import QuantumCircuit, ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------------------------------
# 1. CRIAR O CIRCUITO SEM MEDIÇÃO PARA VER O STATEVECTOR
# -----------------------------------------------------
n = 2                     # número de qubits
marked = '11'             # estado que queremos encontrar

qc = QuantumCircuit(n)

# ----- PASSO 0: Estado inicial |00⟩ -----
display(Markdown("## **Estado inicial**"))
display(qc.draw('mpl'))
display(Markdown("**Statevector:** |00⟩"))

# ----- PASSO 1: Superposição uniforme com portas H -----
qc.h(0)
qc.h(1)
display(Markdown("## **Passo 1 - Aplicar H em todos os qubits**"))
display(qc.draw('mpl'))

# Simular para ver o statevector
sim = AerSimulator(method='statevector')
qc_temp = qc.copy()
qc_temp.save_statevector()
result = sim.run(qc_temp).result()
state = result.get_statevector()
state_array = np.array(state)
display(Markdown(f"**Statevector:** `{np.array2string(state_array, precision=4, suppress_small=True)}`"))
display(Markdown("Ou seja: (|00⟩ + |01⟩ + |10⟩ + |11⟩)/2"))

# ----- PASSO 2: Oracle (marca o estado |11⟩ com fase -1) -----
# Para marcar |11⟩, aplicamos portas X nos qubits que são '0' no estado alvo.
# Como '11' tem apenas 1s, nenhum X é necessário.
# Depois aplicamos uma porta CZ (Z controlada) entre os dois qubits.
# A CZ aplica fase -1 somente quando ambos os qubits estão em |1⟩.
qc.cz(0, 1)
display(Markdown("## **Passo 2 - Oracle (marcar |11⟩ com fase -1)**"))
display(Markdown("Porta CZ entre q0 e q1. Só o estado |11⟩ ganha fase -1."))
display(qc.draw('mpl'))

qc_temp = qc.copy()
qc_temp.save_statevector()
result = sim.run(qc_temp).result()
state = result.get_statevector()
state_array = np.array(state)
display(Markdown(f"**Statevector:** `{np.array2string(state_array, precision=4, suppress_small=True)}`"))
display(Markdown("Agora: (|00⟩ + |01⟩ + |10⟩ **-** |11⟩)/2"))

# ----- PASSO 3: Diffusor (reflexão em torno da média) -----
# O diffusor é composto por: H, X, CZ, X, H (tudo em todos os qubits)
# Para 2 qubits, a sequência é:
qc.h([0,1])
qc.x([0,1])
qc.h(1)          # H no qubit 1 para transformar CZ em MCZ-like
qc.cz(0,1)
qc.h(1)
qc.x([0,1])
qc.h([0,1])
display(Markdown("## **Passo 3 - Diffusor (amplifica o estado marcado)**"))
display(Markdown("Sequência: H, X, H (no último), CZ, H (no último), X, H"))
display(qc.draw('mpl'))

qc_temp = qc.copy()
qc_temp.save_statevector()
result = sim.run(qc_temp).result()
state = result.get_statevector()
state_array = np.array(state)
display(Markdown(f"**Statevector final:** `{np.array2string(state_array, precision=4, suppress_small=True)}`"))
display(Markdown("O estado |11⟩ agora tem amplitude próxima de 1 (probabilidade ≈ 100%)."))

```

```python
# =====================================================
# EXEMPLO DE IMPLEMENTAÇÃO - Banco de dados com 4 itens
# =====================================================
# Aproveitando o circuito 'qc' da célula anterior (já com Oracle e Diffusor aplicados)
# Estado final antes da medição: amplitude concentrada no estado marcado |11⟩

# 1. Definir os itens do banco de dados
itens = ['00', '01', '10', '11']
display(Markdown("# **Exemplo: Banco de dados com 4 itens**"))
display(Markdown(f"**Itens:** {itens}"))
display(Markdown(f"**Item procurado:** {marked}"))

# 2. Evolução do estado (já mostrada passo a passo na célula anterior)
display(Markdown("## **Evolução do estado quântico**"))
display(Markdown("""
- **Início:** |00⟩  
- **Superposição:** (|00⟩+|01⟩+|10⟩+|11⟩)/2  
- **Após Oracle:** fase -1 no item |11⟩ → (|00⟩+|01⟩+|10⟩-|11⟩)/2  
- **Após Diffusor:** amplitude do item |11⟩ é amplificada para ≈ 1, outros itens ≈ 0  
"""))

# 3. Medir o circuito já existente
from qiskit import ClassicalRegister   # só se não tiver sido importado na célula anterior (mas já foi)
cr = ClassicalRegister(n)
qc_meas = qc.copy()
qc_meas.add_register(cr)
qc_meas.measure(range(n), range(n))

display(Markdown("## **Circuito com medição**"))
display(qc_meas.draw('mpl'))

# Executar medições (usando o mesmo simulador, mas sem 'statevector' porque vamos medir)
sim_meas = AerSimulator()
result_meas = sim_meas.run(qc_meas, shots=1024).result()
counts = result_meas.get_counts()

# Verificar se o dicionário counts tem dados
print("Dicionário counts:", counts)

# Se estiver vazio, algo deu errado na simulação
# Se tiver dados, vamos plotar de outra forma
if counts:
    import matplotlib.pyplot as plt
    plt.figure()
    plt.bar(counts.keys(), counts.values())
    plt.xlabel('Estados')
    plt.ylabel('Contagens')
    plt.title('Resultado do Algoritmo de Grover')
    plt.show()
else:
    print("Nenhum resultado encontrado. Verifique o circuito de medição.")
```

```python
# =====================================================
# IMPLEMENTAÇÃO COM RUÍDO E OPERADOR DENSIDADE (CORRIGIDO)
# =====================================================
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError
import numpy as np

# ---- 1. Definir o modelo de ruído ----
noise_model = NoiseModel()

# Erro depolarizante em portas de 1 qubit (probabilidade 1%)
p1 = 0.10   # 10%
p2 = 0.30   # 30%
prob_meas = 0.20   # 20%
depol_1 = depolarizing_error(p1, 1)
# Erro depolarizante em portas de 2 qubits (probabilidade 5%)
depol_2 = depolarizing_error(p2, 2)

# Adicionar erros usando strings vazias ('') para aplicar a todas as portas do circuito
noise_model.add_all_qubit_quantum_error(depol_1, '')  # Erro para portas de 1 qubit
noise_model.add_all_qubit_quantum_error(depol_2, '')  # Erro para portas de 2 qubits

# Erro de leitura (2% de chance de ler 0 como 1 e vice-versa)
prob_meas = prob_meas
readout_error = ReadoutError([[1-prob_meas, prob_meas], [prob_meas, 1-prob_meas]])

# ✅ CORREÇÃO: Adicionar o erro de leitura separadamente em cada qubit
for qubit in [0, 1]:
    noise_model.add_readout_error(readout_error, [qubit])

display(Markdown("## **Modelo de ruído configurado**"))
display(Markdown(f"- Erro depolarizante 1-qubit: {p1:.0%}"))
display(Markdown(f"- Erro depolarizante 2-qubit: {p2:.0%}"))
display(Markdown(f"- Erro de leitura: {prob_meas:.0%}"))

# ---- 2. Criar uma cópia do circuito e adicionar medição ----
qc_noisy = qc.copy()                # reaproveita o circuito já construído
cr = ClassicalRegister(n)
qc_noisy.add_register(cr)
qc_noisy.measure(range(n), range(n))

# ---- 3. Simular com ruído (operador densidade) ----
sim_noisy = AerSimulator(method='density_matrix', noise_model=noise_model)
result_noisy = sim_noisy.run(qc_noisy, shots=8192).result()
counts_noisy = result_noisy.get_counts()

# ---- 4. Exibir resultados ----
import matplotlib.pyplot as plt
if counts_noisy:
    plt.bar(counts_noisy.keys(), counts_noisy.values())
    plt.xlabel('Estados')
    plt.ylabel('Contagens')
    plt.title('Grover com ruído')
    plt.show()
else:
    print("counts_noisy está vazio. Verifique a simulação.")

prob_success_noisy = counts_noisy.get(marked, 0) / 8192
display(Markdown(f"**Probabilidade de encontrar o item `{marked}` com ruído:** {prob_success_noisy:.2%}"))

# ---- 5. Comparar com o caso ideal (sem ruído) ----
sim_ideal = AerSimulator()
result_ideal = sim_ideal.run(qc_noisy, shots=8192).result()
counts_ideal = result_ideal.get_counts()
prob_success_ideal = counts_ideal.get(marked, 0) / 8192

display(Markdown("## **Comparação com o caso ideal (sem ruído)**"))
display(Markdown(f"- **Sucesso ideal:** {prob_success_ideal:.2%}"))
display(Markdown(f"- **Sucesso com ruído:** {prob_success_noisy:.2%}"))
display(Markdown(f"- **Degradação:** {prob_success_ideal - prob_success_noisy:.2%}"))

# ---- 6. Análise da matriz densidade (pureza) ----
qc_dm = qc.copy()
qc_dm.save_density_matrix(label='final_rho')
result_dm = sim_noisy.run(qc_dm).result()
rho_final = result_dm.data()['final_rho']
# Converte para array numpy se necessário
rho_array = np.array(rho_final)
# Calcula a pureza como Tr(ρ²)
purity = np.trace(rho_array @ rho_array).real
display(Markdown(f"**Pureza da matriz densidade final:** {purity:.4f} (1 = estado puro, <1 = misto)"))
display(Markdown("A pureza menor que 1 confirma a presença de decoerência."))
```

```python
# =====================================================
# EXEMPLO DE IMPLEMENTAÇÃO COM RUÍDO
# Banco de dados com 4 itens (2 qubits) - Estado marcado: |11⟩
# Usando operador densidade para modelar o ruído
# =====================================================

from qiskit import ClassicalRegister
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError
from qiskit.visualization import plot_histogram
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Configurar modelo de ruído (valores mais perceptíveis) ---
noise_model = NoiseModel()

# Erros depolarizantes (valores aumentados para ver efeito)
p1 = 0.08      # 8% de erro em portas de 1 qubit
p2 = 0.20      # 20% de erro em portas de 2 qubits
depol_1 = depolarizing_error(p1, 1)
depol_2 = depolarizing_error(p2, 2)

# Aplicar erros a todas as portas do circuito
noise_model.add_all_qubit_quantum_error(depol_1, '')
noise_model.add_all_qubit_quantum_error(depol_2, '')

# Erro de leitura (10% de chance de ler errado)
prob_meas = 0.10
readout_error = ReadoutError([[1-prob_meas, prob_meas], [prob_meas, 1-prob_meas]])
for qubit in range(n):
    noise_model.add_readout_error(readout_error, [qubit])

display(Markdown("## **Modelo de ruído configurado para o exemplo**"))
display(Markdown(f"- Erro depolarizante (1-qubit): {p1:.0%}"))
display(Markdown(f"- Erro depolarizante (2-qubits): {p2:.0%}"))
display(Markdown(f"- Erro de leitura: {prob_meas:.0%}"))

# --- 2. Preparar o circuito com medição (reutilizando qc da célula anterior) ---
qc_noisy = qc.copy()                # qc já contém H, Oracle, Diffusor
cr = ClassicalRegister(n)
qc_noisy.add_register(cr)
qc_noisy.measure(range(n), range(n))

display(Markdown("## **Circuito de Grover com medição (mesmo da versão sem ruído)**"))
display(qc_noisy.draw('mpl'))

# --- 3. Simular com ruído (operador densidade) ---
sim_noisy = AerSimulator(method='density_matrix', noise_model=noise_model)
result_noisy = sim_noisy.run(qc_noisy, shots=8192).result()
counts_noisy = result_noisy.get_counts()

# --- 4. Resultados ---
plt.bar(counts_noisy.keys(), counts_noisy.values())
plt.xlabel('Estados')
plt.ylabel('Contagens')
plt.title('Grover com ruído (8192 shots)')
plt.show()

prob_success_noisy = counts_noisy.get(marked, 0) / 8192
display(Markdown(f"**Probabilidade de encontrar o item `{marked}` com ruído:** {prob_success_noisy:.2%}"))

# --- 5. Comparação com o caso ideal (sem ruído) ---
# Executar o mesmo circuito sem ruído para comparação
sim_ideal = AerSimulator()
result_ideal = sim_ideal.run(qc_noisy, shots=8192).result()
counts_ideal = result_ideal.get_counts()
prob_success_ideal = counts_ideal.get(marked, 0) / 8192

display(Markdown("## **Comparação com o caso ideal (sem ruído)**"))
display(Markdown(f"- **Sucesso ideal:** {prob_success_ideal:.2%}"))
display(Markdown(f"- **Sucesso com ruído:** {prob_success_noisy:.2%}"))
display(Markdown(f"- **Degradação:** {prob_success_ideal - prob_success_noisy:.2%}"))

# --- 6. Análise da matriz densidade (pureza) ---
# Para mostrar como o operador densidade captura o ruído
qc_dm = qc.copy()
qc_dm.save_density_matrix(label='final_rho')
result_dm = sim_noisy.run(qc_dm).result()
rho_final = result_dm.data()['final_rho']
rho_array = np.array(rho_final)
purity = np.trace(rho_array @ rho_array).real
display(Markdown("## **Análise do operador densidade**"))
display(Markdown(f"**Pureza da matriz densidade final:** {purity:.4f}"))
display(Markdown("*(Pureza = 1 indica estado puro (sem ruído); pureza < 1 indica estado misto devido ao ruído)*"))

# --- 7. Descrição qualitativa (explicação) ---
display(Markdown("## **Como o operador densidade lida com o ruído e consequências**"))
display(Markdown("""
**Mecanismo:**  
- O operador densidade $\\rho$ representa um ensemble de estados quânticos. Quando há ruído, o estado puro $|\\psi\\rangle$ evolui para um estado misto $\\rho = \\sum_k p_k |\\psi_k\\rangle\\langle\\psi_k|$.  
- Os canais de ruído (depolarização, erro de leitura) são implementados como operadores de Kraus que transformam $\\rho$ de forma linear e preservam o traço.  
- O simulador `density_matrix` do Qiskit Aer aplica essas transformações a cada porta e medição.

**Consequências no resultado:**  
- A probabilidade de sucesso do algoritmo de Grover diminui (neste exemplo, de {prob_success_ideal:.0%} para {prob_success_noisy:.0%}).  
- O histograma mostra contagens não nulas em estados errados (`00`, `01`, `10`).  
- A pureza final < 1 indica perda de coerência: as superposições se tornam misturas estatísticas, reduzindo a amplificação do estado marcado.  
- O ruído afeta especialmente o diffusor, que depende de interferência construtiva – a decoerência degrada essa interferência.
""".format(prob_success_ideal=prob_success_ideal*100, prob_success_noisy=prob_success_noisy*100)))
```
